```python
#celula responsável por obter os pontos de máximo da onda do eixo aY()

from scipy.signal import argrelextrema
import numpy as np

df_temp = df[['aY ()']]
ilocs_min = argrelextrema(df_temp['aY ()'].values, np.less_equal, order=10)[0]
ilocs_max = argrelextrema(df_temp['aY ()'].values, np.greater_equal, order=10)[0]

df_temp['aY ()'].plot(figsize=(30,15))
# filter prices that are peaks and plot them differently to be visable on the plot
df_temp.iloc[ilocs_max].plot(style='.', lw=10, color='red', marker="v");
df_temp.iloc[ilocs_min].plot(style='.', lw=10, color='green', marker="^");
```

```python
#Célula responsável por criar coluna na tabela de dados com dados de validação gerados com base nos pontos máximos de cada onda 

df_temp['weekly_max'] = False
df_temp['weekly_min'] = False
df_temp['sinal ()'] = 0
df_temp.loc[df_temp.iloc[ilocs_min].index, 'weekly_min'] = True
df_temp.loc[df_temp.iloc[ilocs_max].index, 'weekly_max'] = True
df_temp.loc[df_temp.iloc[ilocs_max].index, 'sinal ()'] = 200
print(df_temp[df_temp['weekly_min'] | df_temp['weekly_max']])
#df2.loc[df2['weekly_min'] == True, 'sinal ()'] = 1
#df2.loc[df2['aY ()'] >= 0, 'sinal ()'] = 0
a = 10
for i in range(len(df_temp)):
   
  #print(a)
  
  if df_temp.iloc[i]['sinal ()'] == 200 and a>10 and df_temp.iloc[i]['aY ()'] > 0 :

    df_temp.iloc[i-10:i+5]['sinal ()']=200
    
    a = 0
    #print(i)
  elif df_temp.iloc[i]['sinal ()'] == 200 and a>10 and df_temp.iloc[i]['aY ()'] < 0:
    df_temp.iloc[i-10:i+5]['sinal ()']=0

  a+=1


```

```python
#verificação da onda

df_temp.plot(figsize=(30,15), alpha=.3)
df_temp['aY ()'].plot
df_temp['sinal ()'].plot
#df2[df2['weekly_max']].plot(style='.', lw=10, color='red', marker="v");
#df2[df2['weekly_min']].plot(style='.', lw=10, color='green', marker="^");
plt.xlim([10000,15000])
```

```python
#remoção de colunas desnecessárias  

#df_temp.loc[df_temp['sinal ()'] == 0, 'sinal ()'] = 'OFF'
#df_temp.loc[df_temp['sinal ()'] == 1000, 'sinal ()'] = 'ON'
df_temp1 = df_temp

df_temp1 = df_temp1.drop('aY ()', 1)
df_temp1 = df_temp1.drop('weekly_max', 1)
df_temp1 = df_temp1.drop('weekly_min', 1)
df_temp1
```
